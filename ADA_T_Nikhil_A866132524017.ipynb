#Aim

Implement Quick Sort and analyze its time complexity.

#Theory

Quick Sort is a Divide and Conquer algorithm. It selects a pivot element and partitions the array such that elements smaller than the pivot are placed before it and larger elements after it.

#Time Complexity
| Case    | Complexity |
| ------- | ---------- |
| Best    | O(n log n) |
| Average | O(n log n) |
| Worst   | O(n²)      |

#Space Complexity
O(log n)

```python
# --------------------------------------------------------
# Experiment 1: Quick Sort
# Program 1: Quick Sort using Divide and Conquer
# --------------------------------------------------------

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1


def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)

        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)


arr = list(map(int, input("Enter elements: ").split()))

quick_sort(arr, 0, len(arr)-1)

print("Sorted Array:", arr)
```

#Aim

To implement the Heap Sort algorithm using a Max Heap and verify its time complexity for different input sizes.

#Theory

Heap Sort is a comparison-based sorting algorithm that uses the Binary Heap data structure. It first builds a Max Heap from the given array, where the largest element is always at the root. The root element is then swapped with the last element, the heap size is reduced, and the heap property is restored. This process continues until the array is completely sorted.

Unlike Quick Sort, Heap Sort guarantees O(n log n) time complexity in the best, average, and worst cases.

#Algorithm
->Build a Max Heap from the input array.

->Swap the root (largest element) with the last element.

->Reduce the heap size by one.

->Heapify the root to maintain the Max Heap property.

->Repeat Steps 2–4 until only one element remains.

->The array is now sorted in ascending order.
#Time Complexity

| Case         | Complexity     |
| ------------ | -------------- |
| Best Case    | **O(n log n)** |
| Average Case | **O(n log n)** |
| Worst Case   | **O(n log n)** |

#Space Complexity
O(1)

```python
# --------------------------------------------------------
# Experiment 2: Heap Sort
# Program 1: Heap Sort using Max Heap
# --------------------------------------------------------

def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)


def heap_sort(arr):
    n = len(arr)

    # Build Max Heap
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Extract elements one by one
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, i, 0)


arr = list(map(int, input("Enter the elements: ").split()))

heap_sort(arr)

print("Sorted Array:", arr)
```

->To implement the Merge Sort algorithm using the Divide and Conquer technique and verify its time complexity for different input sizes.

#Theory

Merge Sort is a stable and efficient sorting algorithm based on the Divide and Conquer strategy. It repeatedly divides the given array into two halves until each sub-array contains only one element. The sub-arrays are then merged in sorted order to produce the final sorted array.

Merge Sort performs consistently well regardless of the input arrangement and is widely used for sorting large datasets.

#Algorithm
->Divide the array into two equal halves.

->Recursively divide each half until only one element remains.

->Merge the divided arrays by comparing their elements.

->Continue merging until the complete array is obtained.

->Display the sorted array.
#Time Complexity
| Case         | Complexity     |
| ------------ | -------------- |
| Best Case    | **O(n log n)** |
| Average Case | **O(n log n)** |
| Worst Case   | **O(n log n)** |
#Space Complexity

O(n)

```python
# --------------------------------------------------------
# Experiment 3: Merge Sort
# Program 1: Merge Sort using Divide and Conquer
# --------------------------------------------------------

def merge_sort(arr):
    if len(arr) > 1:

        mid = len(arr) // 2

        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

        while i < len(left) and j < len(right):

            if left[i] <= right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1

            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1


arr = list(map(int, input("Enter the elements: ").split()))

merge_sort(arr)

print("Sorted Array:", arr)
```

#EXPERIMENT 4
#Title

Implementation of 0/1 Knapsack using Dynamic Programming and Fractional Knapsack using Greedy Method

#Aim

To implement the 0/1 Knapsack Problem using Dynamic Programming and the Fractional Knapsack Problem using the Greedy Approach, and compare their solutions.

#Theory

The Knapsack Problem is a famous optimization problem in Algorithm Design. It aims to maximize the total profit of selected items without exceeding the capacity of the knapsack.

0/1 Knapsack: Each item can either be selected completely or not selected at all. Dynamic Programming is used to obtain the optimal solution.
Fractional Knapsack: Items can be divided into fractions. A Greedy strategy is used by selecting items based on the highest profit-to-weight ratio.

Dynamic Programming guarantees the optimal solution for the 0/1 Knapsack problem, whereas the Greedy method provides the optimal solution for the Fractional Knapsack problem.

#Algorithm (0/1 Knapsack)

->Read the weights and values of all items.

->Create a DP table of size (n+1) × (W+1).

->Initialize the first row and first column to zero.

->For each item, check whether it can be included.

->Store the maximum profit obtained.

->Display the maximum profit.
#Time Complexity
| Case         | Complexity   |
| ------------ | ------------ |
| Best Case    | **O(n × W)** |
| Average Case | **O(n × W)** |
| Worst Case   | **O(n × W)** |

#Space Complexity

O(n × W)

```python
# --------------------------------------------------------
# Experiment 4: 0/1 Knapsack
# Program 1: Dynamic Programming Approach
# --------------------------------------------------------

def knapsack(W, wt, val, n):

    dp = [[0 for _ in range(W + 1)] for _ in range(n + 1)]

    for i in range(n + 1):
        for w in range(W + 1):

            if i == 0 or w == 0:
                dp[i][w] = 0

            elif wt[i - 1] <= w:
                dp[i][w] = max(val[i - 1] + dp[i - 1][w - wt[i - 1]],
                               dp[i - 1][w])

            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][W]


values = [60, 100, 120]
weights = [10, 20, 30]
capacity = 50

print("Maximum Profit =", knapsack(capacity, weights, values, len(values)))
```

```python
# --------------------------------------------------------
# Experiment 4: Fractional Knapsack
# Program 2: Greedy Method
# --------------------------------------------------------

def fractional_knapsack(capacity, items):

    items.sort(key=lambda x: x[1] / x[0], reverse=True)

    total_profit = 0

    for weight, value in items:

        if capacity >= weight:

            capacity -= weight
            total_profit += value

        else:

            total_profit += value * (capacity / weight)
            break

    return total_profit


items = [(10, 60), (20, 100), (30, 120)]

capacity = 50

print("Maximum Profit =", fractional_knapsack(capacity, items))
```

#Observation
Dynamic Programming provides the optimal solution for the 0/1 Knapsack problem.


Greedy Method provides the optimal solution for the Fractional Knapsack problem.


Fractional Knapsack usually gives a higher profit because items can be divided.
#Conclusion

The 0/1 Knapsack Problem was successfully implemented using Dynamic Programming, while the Fractional Knapsack Problem was implemented using the Greedy approach. The results verified that Dynamic Programming is suitable for indivisible items, whereas the Greedy strategy is optimal when fractional selection is allowed.

#Title

Implementation of Kruskal's Algorithm to Find the Minimum Spanning Tree (MST)

#Aim

To implement Kruskal's Algorithm for finding the Minimum Spanning Tree (MST) of a weighted undirected graph and verify its correctness.

#Theory

A Minimum Spanning Tree (MST) is a subset of edges in a connected, weighted, undirected graph that connects all vertices with the minimum possible total edge weight and without forming any cycles.



Kruskal's Algorithm is a Greedy algorithm that builds the MST by selecting the edges with the smallest weights one by one while ensuring that no cycle is formed. It uses the Disjoint Set (Union-Find) data structure for cycle detection.

#Algorithm

->Sort all edges in increasing order of their weights.

->Initialize each vertex as a separate set.

->Select the edge with the minimum weight.

->Check whether including the edge forms a cycle.

->If no cycle is formed, include the edge in the MST.

->Repeat until (V − 1) edges are selected.
#Time Complexity
| Case         | Complexity     |
| ------------ | -------------- |
| Best Case    | **O(E log E)** |
| Average Case | **O(E log E)** |
| Worst Case   | **O(E log E)** |

#Space Complexity

O(V)

```python
# --------------------------------------------------------
# Experiment 5: Kruskal's Algorithm
# Program 1: Minimum Spanning Tree using Kruskal's Algorithm
# --------------------------------------------------------

class Graph:

    def __init__(self, vertices):
        self.V = vertices
        self.graph = []

    def add_edge(self, u, v, w):
        self.graph.append([u, v, w])

    def find(self, parent, i):
        if parent[i] == i:
            return i
        return self.find(parent, parent[i])

    def union(self, parent, rank, x, y):

        xroot = self.find(parent, x)
        yroot = self.find(parent, y)

        if rank[xroot] < rank[yroot]:
            parent[xroot] = yroot

        elif rank[xroot] > rank[yroot]:
            parent[yroot] = xroot

        else:
            parent[yroot] = xroot
            rank[xroot] += 1

    def kruskal(self):

        result = []
        i = 0
        e = 0

        self.graph = sorted(self.graph, key=lambda item: item[2])

        parent = []
        rank = []

        for node in range(self.V):
            parent.append(node)
            rank.append(0)

        while e < self.V - 1:

            u, v, w = self.graph[i]
            i += 1

            x = self.find(parent, u)
            y = self.find(parent, v)

            if x != y:
                e += 1
                result.append([u, v, w])
                self.union(parent, rank, x, y)

        print("Edges in Minimum Spanning Tree:\n")

        total = 0

        for u, v, w in result:
            print(f"{u} -- {v} == {w}")
            total += w

        print("\nTotal Weight =", total)


g = Graph(4)

g.add_edge(0, 1, 10)
g.add_edge(0, 2, 6)
g.add_edge(0, 3, 5)
g.add_edge(1, 3, 15)
g.add_edge(2, 3, 4)

g.kruskal()
```

#Title

Implementation of Prim's Algorithm to Find the Minimum Spanning Tree (MST)

#Aim

To implement Prim's Algorithm for finding the Minimum Spanning Tree (MST) of a weighted undirected graph and verify its correctness.

#Theory

A Minimum Spanning Tree (MST) is a tree that connects all the vertices of a connected, weighted, undirected graph with the minimum possible total edge weight.

Prim's Algorithm is a Greedy algorithm that starts from any vertex and repeatedly adds the minimum-weight edge that connects a visited vertex to an unvisited vertex until all vertices are included in the MST.

Unlike Kruskal's Algorithm, Prim's Algorithm grows a single tree by selecting the nearest vertex at every step.

#Algorithm
->Select any vertex as the starting vertex.

->Mark it as visited.

->Find the minimum-weight edge connecting a visited vertex to an unvisited vertex.

->Add the selected edge to the MST.

->Mark the new vertex as visited.

->Repeat until all vertices are included in the MST.
#Time Complexity
| Case         | Complexity |
| ------------ | ---------- |
| Best Case    | **O(V²)**  |
| Average Case | **O(V²)**  |
| Worst Case   | **O(V²)**  |
#Space Complexity

O(V²)

```python
# --------------------------------------------------------
# Experiment 6: Prim's Algorithm
# Program 1: Minimum Spanning Tree using Prim's Algorithm
# --------------------------------------------------------

INF = 9999999

V = 5

graph = [
    [0, 9, 75, 0, 0],
    [9, 0, 95, 19, 42],
    [75, 95, 0, 51, 66],
    [0, 19, 51, 0, 31],
    [0, 42, 66, 31, 0]
]

selected = [False] * V

selected[0] = True

print("Edge : Weight\n")

edge_count = 0
total_cost = 0

while edge_count < V - 1:

    minimum = INF
    x = 0
    y = 0

    for i in range(V):

        if selected[i]:

            for j in range(V):

                if (not selected[j]) and graph[i][j]:

                    if minimum > graph[i][j]:

                        minimum = graph[i][j]
                        x = i
                        y = j

    print(x, "-", y, ":", graph[x][y])

    total_cost += graph[x][y]

    selected[y] = True

    edge_count += 1

print("\nTotal Cost =", total_cost)
```

#EXPERIMENT 7
#Title

Implementation of Travelling Salesman Problem (TSP) using Dynamic Programming

#Aim

To implement the Travelling Salesman Problem (TSP) using Dynamic Programming and determine the minimum cost required to visit all cities exactly once and return to the starting city.

#Theory

The Travelling Salesman Problem (TSP) is a well-known optimization problem in computer science. Given a set of cities and the distances between every pair of cities, the objective is to find the shortest possible route that visits each city exactly once and returns to the starting city.

Dynamic Programming improves efficiency by storing the results of previously solved subproblems, thereby avoiding redundant computations.

Applications of TSP include route planning, logistics, circuit board design, and delivery optimization.

#Algorithm
->Represent the cities using a distance matrix.

->Start from the source city.

->Mark the visited cities using a bitmask.

->Recursively visit each unvisited city.

->Store intermediate results using Dynamic Programming.

->Return the minimum cost after visiting all cities.
#Time Complexity
| Case         | Complexity     |
| ------------ | -------------- |
| Best Case    | **O(n² × 2ⁿ)** |
| Average Case | **O(n² × 2ⁿ)** |
| Worst Case   | **O(n² × 2ⁿ)** |

#Space Complexity

O(n × 2ⁿ)

```python
# --------------------------------------------------------
# Experiment 7: Travelling Salesman Problem
# Program 1: Dynamic Programming using Memoization
# --------------------------------------------------------

from functools import lru_cache

graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

n = len(graph)

@lru_cache(None)
def tsp(mask, pos):

    if mask == (1 << n) - 1:
        return graph[pos][0]

    ans = float('inf')

    for city in range(n):
        if mask & (1 << city) == 0:
            ans = min(ans,
                      graph[pos][city] +
                      tsp(mask | (1 << city), city))

    return ans


print("Minimum Cost =", tsp(1, 0))
```

```python
# --------------------------------------------------------
# Experiment 7: Travelling Salesman Problem
# Program 2: Brute Force Approach
# --------------------------------------------------------

from itertools import permutations

graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

n = len(graph)

cities = list(range(1, n))

minimum_cost = float('inf')

for path in permutations(cities):

    cost = 0
    current = 0

    for city in path:
        cost += graph[current][city]
        current = city

    cost += graph[current][0]

    minimum_cost = min(minimum_cost, cost)

print("Minimum Cost =", minimum_cost)
```

#Observation
The Dynamic Programming approach avoids repeated calculations by storing intermediate results.


The Brute Force method checks every possible route, making it inefficient for larger numbers of cities.


Dynamic Programming significantly reduces execution time compared to the Brute Force approach.
#Conclusion

The Travelling Salesman Problem was successfully implemented using Dynamic Programming and Brute Force methods. The Dynamic Programming approach produced the optimal solution with improved efficiency by utilizing memoization, making it more suitable for larger problem sizes.

#EXPERIMENT 8
#Title

Implementation of N-Queen Problem using Backtracking

#Aim

To implement the N-Queen Problem using the Backtracking Algorithm and find all possible solutions for placing N queens on an N × N chessboard such that no two queens attack each other.

#Theory

The N-Queen Problem is a classic problem in Backtracking. The objective is to place N queens on an N × N chessboard so that no two queens share the same row, column, or diagonal.

Backtracking is a recursive problem-solving technique that builds a solution incrementally. If a partial solution violates the problem constraints, it backtracks and tries another possibility.

The N-Queen problem demonstrates the power of recursion and pruning in solving combinatorial optimization problems.

#Algorithm
->Start placing queens from the first column.

->Check whether the current position is safe.

->If safe, place the queen.

->Recursively place queens in the next column.

->If no safe position exists, remove the previously placed queen (backtrack).

->Repeat until all queens are placed successfully.
#Time Complexity
| Case         | Complexity |
| ------------ | ---------- |
| Best Case    | **O(N!)**  |
| Average Case | **O(N!)**  |
| Worst Case   | **O(N!)**  |
#Space Complexity

O(N²)

```python
# --------------------------------------------------------
# Experiment 8: N-Queen Problem
# Program 1: N-Queen using Backtracking
# --------------------------------------------------------

N = 4

board = [[0 for _ in range(N)] for _ in range(N)]

def is_safe(board, row, col):

    for i in range(col):
        if board[row][i] == 1:
            return False

    i = row
    j = col

    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1

    i = row
    j = col

    while i < N and j >= 0:
        if board[i][j] == 1:
            return False
        i += 1
        j -= 1

    return True


def solve(board, col):

    if col >= N:
        return True

    for i in range(N):

        if is_safe(board, i, col):

            board[i][col] = 1

            if solve(board, col + 1):
                return True

            board[i][col] = 0

    return False


if solve(board, 0):

    print("Solution Matrix:\n")

    for row in board:
        print(row)

else:

    print("No Solution Exists")
```

#Title

Implementation of Strassen's Matrix Multiplication Algorithm

#Aim

To implement Strassen's Matrix Multiplication Algorithm and compare it with the conventional matrix multiplication method for efficient multiplication of square matrices.

#Theory

Matrix multiplication is a fundamental operation in computer science and engineering. The conventional method requires O(n³) time complexity. Strassen's Algorithm, developed by Volker Strassen, is a Divide and Conquer algorithm that reduces the number of recursive multiplications from 8 to 7, thereby improving efficiency.

The matrices are recursively divided into four equal-sized submatrices, and seven intermediate products are computed to obtain the final result.

Strassen's algorithm is efficient for large square matrices and is widely used in scientific computing and image processing.

#Algorithm
->Divide each matrix into four equal submatrices.

->Compute seven intermediate matrices (P₁ to P₇).

->Calculate the four quadrants of the resultant matrix.

->Combine the four quadrants into a single matrix.

->Display the resultant matrix.
#Time Complexity
| Case         | Complexity    |
| ------------ | ------------- |
| Best Case    | **O(n^2.81)** |
| Average Case | **O(n^2.81)** |
| Worst Case   | **O(n^2.81)** |
#Space Complexity

O(n²)

```python
# --------------------------------------------------------
# Experiment 9: Strassen's Matrix Multiplication
# Program 1: Strassen's Algorithm
# --------------------------------------------------------

import numpy as np

def strassen(A, B):

    if len(A) == 1:
        return A * B

    n = len(A)

    mid = n // 2

    A11 = A[:mid, :mid]
    A12 = A[:mid, mid:]
    A21 = A[mid:, :mid]
    A22 = A[mid:, mid:]

    B11 = B[:mid, :mid]
    B12 = B[:mid, mid:]
    B21 = B[mid:, :mid]
    B22 = B[mid:, mid:]

    P1 = strassen(A11 + A22, B11 + B22)
    P2 = strassen(A21 + A22, B11)
    P3 = strassen(A11, B12 - B22)
    P4 = strassen(A22, B21 - B11)
    P5 = strassen(A11 + A12, B22)
    P6 = strassen(A21 - A11, B11 + B12)
    P7 = strassen(A12 - A22, B21 + B22)

    C11 = P1 + P4 - P5 + P7
    C12 = P3 + P5
    C21 = P2 + P4
    C22 = P1 - P2 + P3 + P6

    top = np.hstack((C11, C12))
    bottom = np.hstack((C21, C22))

    return np.vstack((top, bottom))


A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:\n", A)

print("\nMatrix B:\n", B)

print("\nResultant Matrix:\n")

print(strassen(A, B))
```

Implementation of Ford-Fulkerson Algorithm to Find the Maximum Flow in a Flow Network

#Aim

To implement the Ford-Fulkerson Algorithm for finding the Maximum Flow in a flow network and determine the maximum possible flow from the source vertex to the sink vertex.

#Theory

The Maximum Flow Problem is a fundamental problem in graph theory that determines the greatest possible flow from a source vertex to a sink vertex in a flow network without violating the capacity constraints of the edges.

The Ford-Fulkerson Algorithm solves this problem by repeatedly finding an augmenting path from the source to the sink using Breadth-First Search (BFS) or Depth-First Search (DFS). The flow along the path is increased until no more augmenting paths exist.

This algorithm is widely used in transportation, communication networks, water supply systems, and network routing.

#Algorithm
->Initialize the maximum flow as zero.

->Find an augmenting path from the source to the sink using BFS.

->Determine the minimum residual capacity (bottleneck capacity) along the path.

->Increase the flow by the bottleneck capacity.

->Update the residual graph.

->Repeat Steps 2–5 until no augmenting path exists.

->Display the maximum flow.
#Time Complexity
| Case         | Complexity |
| ------------ | ---------- |
| Best Case    | **O(VE²)** |
| Average Case | **O(VE²)** |
| Worst Case   | **O(VE²)** |
#Space Complexity

O(V²)

```python
# --------------------------------------------------------
# Experiment 10: Ford-Fulkerson Algorithm
# Program 1: Maximum Flow using BFS (Edmonds-Karp)
# --------------------------------------------------------

from collections import deque

class Graph:

    def __init__(self, graph):
        self.graph = graph
        self.row = len(graph)

    def bfs(self, s, t, parent):

        visited = [False] * self.row
        queue = deque([s])
        visited[s] = True

        while queue:

            u = queue.popleft()

            for ind, val in enumerate(self.graph[u]):

                if not visited[ind] and val > 0:

                    queue.append(ind)
                    visited[ind] = True
                    parent[ind] = u

        return visited[t]

    def ford_fulkerson(self, source, sink):

        parent = [-1] * self.row

        max_flow = 0

        while self.bfs(source, sink, parent):

            path_flow = float("Inf")

            s = sink

            while s != source:

                path_flow = min(path_flow,
                                self.graph[parent[s]][s])
                s = parent[s]

            max_flow += path_flow

            v = sink

            while v != source:

                u = parent[v]
                self.graph[u][v] -= path_flow
                self.graph[v][u] += path_flow
                v = parent[v]

        return max_flow


graph = [
    [0,16,13,0,0,0],
    [0,0,10,12,0,0],
    [0,4,0,0,14,0],
    [0,0,9,0,0,20],
    [0,0,0,7,0,4],
    [0,0,0,0,0,0]
]

g = Graph(graph)

print("Maximum Flow =", g.ford_fulkerson(0,5))
```

#Algorithm

->Update the distances of all adjacent vertices if a shorter path is found.

->Repeat Steps 3–5 until all vertices are visited.

->Display the shortest distance from the source to every other vertex.
#Time Complexity
| Case         | Complexity |
| ------------ | ---------- |
| Best Case    | **O(V²)**  |
| Average Case | **O(V²)**  |
| Worst Case   | **O(V²)**  |
#Space Complexity

```python
# --------------------------------------------------------
# Experiment 11: Dijkstra's Algorithm
# Program 1: Shortest Path using Adjacency Matrix
# --------------------------------------------------------

import sys

V = 6

graph = [
    [0, 4, 2, 0, 0, 0],
    [4, 0, 1, 5, 0, 0],
    [2, 1, 0, 8, 10, 0],
    [0, 5, 8, 0, 2, 6],
    [0, 0, 10, 2, 0, 3],
    [0, 0, 0, 6, 3, 0]
]

distance = [sys.maxsize] * V
visited = [False] * V

distance[0] = 0

for _ in range(V):

    minimum = sys.maxsize
    u = -1

    for i in range(V):
        if not visited[i] and distance[i] < minimum:
            minimum = distance[i]
            u = i

    visited[u] = True

    for v in range(V):
        if (graph[u][v] > 0 and
            not visited[v] and
            distance[v] > distance[u] + graph[u][v]):

            distance[v] = distance[u] + graph[u][v]

print("Vertex\tDistance from Source")

for i in range(V):
    print(i, "\t", distance[i])
```

#Algorithm
->Read the set of integers and the target sum.

->Start with an empty subset.

#Space Complexity

O(n) (Recursion Stack)

```python
# --------------------------------------------------------
# Experiment 13: Subset Sum Problem
# Program 1: Subset Sum using Backtracking
# --------------------------------------------------------

def subset_sum(nums, target, index, current_subset):

    if target == 0:
        print(current_subset)
        return

    if index == len(nums) or target < 0:
        return

    # Include current element
    subset_sum(nums,
               target - nums[index],
               index + 1,
               current_subset + [nums[index]])

    # Exclude current element
    subset_sum(nums,
               target,
               index + 1,
               current_subset)


numbers = [3, 4, 5, 2]

target = 7

print("Subsets with Sum =", target)

subset_sum(numbers, target, 0, [])
```

#Algorithm
